```r
library(MASS)
library(ellipse)
```

```r
mu = c(3,3)
rho = matrix( c(1, .9, .9, 1), nrow=2, ncol=2)
sds = c(1,1)

# mvrnorm takes mu and sigma, the covraiance matrix:
sigma = diag(sds)%*%rho%*%diag(sds)

x = mvrnorm(1000, mu, sigma)
```

```r
par(pty="s")
xlim = extendrange(x[,1:2], f = 0.15)
ylim = xlim
plot(x, xlab="", ylab="", xlim=xlim, ylim=ylim, col="red")
```

```r
#estimate the parameters
xbar = apply(x, 2, mean)
diagS = apply(x,2,sd)
S = cov(x)
R = cor(x)
```

```r
#estimated density contour:
lines(ellipse(R, scale=diagS, centre=xbar, level=.95), type="1", col = "blue")
```
